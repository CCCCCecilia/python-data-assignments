```python
txts_path = ["trade-wars-news1.txt" , "trade-wars-news2.txt" , "trade-wars-news3.txt" , "trade-wars-news4.txt" , "trade-wars-news5.txt"]
all_content = ""
for txt_path in txts_path:
    txt = open(txt_path, "r")
    content = txt.read().lower()
    all_content = all_content+content
    print(content)
```

```python
punc_file = open("punctuation.txt","r")
puncs = punc_file.read().split("\n")
all_content = all_content.replace("’"," ")
all_content = all_content.replace("'"," ")
for punc in puncs:
    all_content = all_content.replace(punc, "")
print(all_content)
```

```python
all_words = str.split(all_content)
keywords = set(all_words)
```

```python
key_fre = {}
for keyword in keywords:
    key_fre.setdefault(keyword, 0)
for word in all_words:
    key_fre[word] = key_fre[word] + 1
print(key_fre)
```

```python
stop_words = open("stop_word.txt", 'r').read().split("\n")
for stop_word in stop_words:
    if key_fre.get(stop_word)!=None:
         key_fre[stop_word] = 0
    
```

```python
sorted_key_fre = sorted(key_fre.items(),key = lambda x:x[1], reverse=True)
print(sorted_key_fre[:15])
```

```python
import csv
csv_file = open("keyword, frequency.csv", "w",newline='')
mywriter = csv.writer(csv_file)
mywriter.writerow(["key", "frequency"])
for i in range(15):
    print(sorted_key_fre[i])
    mywriter.writerow(sorted_key_fre[i])
csv_file.close()
```
